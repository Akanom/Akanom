```python
from random import randint
from plotly.graph_objs import Bar,Layout
from plotly import offline 
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

<h2>Creating a class and initializing the data</h2>

```python
class Play:
    """A class representing a single die"""
    def __init__(self, sides=6):
        """create a six sided dice"""
        self.sides=sides
    def roll(self):
        """return a random value between 1 and number of sides"""
        return randint(1, self.sides)
```

```python
die=Play()#create a D6 die
die_2=Play()#create another dices
results=[]#container to hold rolls
for roll_number in range(10_000):
    result=die.roll()+die_2.roll()
    results.append(result)
    
frequencies=[]#container to hold the number of times each number occur
highest_result=die.sides+die_2.sides
for freq in range(2, highest_result+1):
    frequency= results.count(freq)
    frequencies.append(frequency)
print(frequencies)
```

<h2>Plotting Barchart</h2>

```python
x_data=list(range(2, highest_result+1))
data=[Bar(x=x_data,y=frequencies)]

x_axis_config={"title":"Dice Results","dtick":1}
y_axis_config={"title":"Dice Frequencies"}
my_layout=Layout(title="Histogram of Rolling Dice 10,000 times",
                xaxis=x_axis_config,yaxis=y_axis_config)
offline.iplot({"data":data, "layout":my_layout},filename="Rolling dice.html")
```

<h2>Rolling Dice of Different size</h2>

```python
die=Play()#create a D6 die
die_2=Play(15)#create another dices of size 20
results=[]#container to hold rolls
for roll_number in range(100_000):
    result=die.roll()+die_2.roll()
    results.append(result)
    
frequencies=[]#container to hold the number of times each number occur
highest_result=die.sides+die_2.sides
for freq in range(2, highest_result+1):
    frequency= results.count(freq)
    frequencies.append(frequency)
print(frequencies)
```

```python
x_data=list(range(2, highest_result+1))
data=[Bar(x=x_data,y=frequencies)]

x_axis_config={"title":"Dice Results","dtick":1}
y_axis_config={"title":"Dice Frequencies"}
my_layout=Layout(title="Histogram of Rolling Dice 10,000 times",
                xaxis=x_axis_config,yaxis=y_axis_config)
offline.iplot({"data":data, "layout":my_layout},filename="Rolling dice.html")
```
